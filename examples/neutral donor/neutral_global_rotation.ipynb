```python
import numpy as np
import matplotlib.pyplot as plt
from psyduck import Spin, SpinComposite
```

```python
nucleus = Spin(7/2)
electron = Spin(1/2)

donor = SpinComposite([nucleus, electron])

donor.global_rotate(np.pi, 'x', spin_idx=0)
donor.global_rotate(np.pi, 'x', spin_idx=1)
```

```python
donor.state
```

```python
donor.Iz(spin_idx=0)
```
